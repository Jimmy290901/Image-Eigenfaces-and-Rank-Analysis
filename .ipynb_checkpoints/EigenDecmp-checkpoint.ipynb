```python
#Eigen Decomposition of Original Matrix
matrix = np.matrix('1 2 3; 4 5 6; 7 8 9')
print(matrix)
eigenvalues, eigenvectors = np.linalg.eig(matrix)
print("Eigenvalues: \n", eigenvalues)
print("Eigenvectors: \n", eigenvectors)
```

```python
#Adding Noise to the original matrix
noise = np.random.normal(0,1,matrix.shape)
noisy_matrix = noise + matrix
print("Noisy Matrix: \n", noisy_matrix)
noisy_eigenvalues, noisy_eigenvectors = np.linalg.eig(noisy_matrix)
print("Noisy Eigenvalues: \n", noisy_eigenvalues)
print("Noisy Eigenvectors: \n", noisy_eigenvectors)
```

```python
def displayImage(image, title="No title"):
    plot.figure()
    plot.imshow(image)
    plot.title(title)
```

```python
from PIL import Image
import matplotlib.pyplot as plot
import numpy as np

sample_img = Image.open('sample_img.jpg')
displayImage(sample_img, 'Sample Image')
# print("Sample Image object:\n", sample_img)


#Extracting Pixels matrix from original image ------------------------------------------
pixelMatrix = np.matrix(list(sample_img.getdata(band=0)), float)
pixelMatrix.shape = (sample_img.size[1], sample_img.size[0])
displayImage(pixelMatrix, 'Band=0 Pixel Matrix')
# --------------------------------------------------------------------------------------
# print("Original image:\n", pixelMatrix)

#Converting to grayscale image
# gray_img = sample_img.convert('LA')
# plot.figure(figsize=(9,6))
# plot.imshow(gray_img)

#Extracting Pixels matrix from Grayscaled image
# pixelMatrix = np.matrix(list(gray_img.getdata(band=0)), float)
# pixelMatrix.shape = (gray_img.size[1], gray_img.size[0])
# print("Grayscale image:\n", pixelMatrix)
#Plotting the grayscale image

#SVD decomposition of original image -----------------------------------------------------
U,S,V = np.linalg.svd(pixelMatrix)
totalComps = U[0].size
# --------------------------------------------------------------------------------------

#Testing to see how the image looks like with partial number of eigen components -------------------
x = 10
y = 100
for comps in range(0, y, x):
    newImgPixelMatrix = np.matrix(U[:,:comps]*np.diag(S[:comps])*np.matrix(V[:comps,:]))
    displayImage(newImgPixelMatrix, "Number of Eigen Components = " + str(comps))
# ---------------------------------------------------------------------------------------------------
```

```python
from sklearn import datasets, decomposition
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plot
import cv2
```

```python
def eigenFaces(facesDataset):
    print("Shape of facesDataset.data: ",facesDataset.data.shape)
    #To access the ith image, we do - facesDataset.images[i]
    print("First 20 images in the dataset:")
    figure = plot.figure(figsize=(8,6))
    for i in range(20):
        img = figure.add_subplot(4,5,i+1, xticks=[], yticks=[])
        img.imshow(facesDataset.images[i], cmap=plot.cm.bone)

    trainset, testset, index_trainset, index_testset = train_test_split(facesDataset.data,facesDataset.target)
    # print(trainset.shape, testset.shape)
    pcaModel = decomposition.PCA(n_components=150)
    pcaModel.fit(trainset)

    plot.figure()
    plot.title("Mean Face Image from the dataset (150 eigen components taken):")
    plot.imshow(pcaModel.mean_.reshape(facesDataset.images[0].shape), cmap=plot.cm.bone)

    figure = plot.figure(figsize=(16,6))
    #Printing the first 50 components of the pcaModel
    plot.title("First 50 PCA components")
    for i in range(50):
        img = figure.add_subplot(5,10,i+1,xticks=[],yticks=[])
        img.imshow(pcaModel.components_[i].reshape(facesDataset.images[0].shape),cmap=plot.cm.bone)
```

```python
#AT&T Olivetti images dataset

facesDataset = datasets.fetch_olivetti_faces() 
#facesDataset contains 40 rows, each of 10 images of a person
#Dimension of each image is 64x64
#To access the entire pixel matrix of all images, we can do facesDataset.data (see below)
eigenFaces(facesDataset)
```

```python
#Scaling down the faces
# type(facesDataset.images)
def resizeImages(facesDataset,x,y):
    resImages = []
    for i in range(facesDataset.data.shape[0]):
        resImages.append(cv2.resize(facesDataset.images[i],(x,y)))
    resImages = np.array(resImages)
    print(resImages.shape)
    scaledImages = np.ndarray(shape=resImages.shape)
    for i in range(resImages.shape[0]):
        for j in range(resImages.shape[1]):
            for k in range(resImages.shape[2]):
                scaledImages[i][j][k] = resImages[i][j][k]
    return scaledImages
# print(cv2.resize(facesDataset.images[i],(200,200)))
# print(sdImages.shape)
# for i in range(sdFacesDataset.data.shape[0]):
#     sdFacesDataset.append(cv2.resize(facesDataset.images[i], (200,200)))
# sdFacesDataset = sklearn.utils.Bunch()
# print(sdFacesDataset.data.shape)
# eigenFaces(sdFacesDataset)
```

```python
# Scaling Up Images
x, y = 200, 200
sdFacesDataset = facesDataset
sdFacesDataset.images = resizeImages(facesDataset,x,y)
sdFacesDataset.data = sdFacesDataset.images.flatten().reshape(facesDataset.data.shape[0],x*y)
print(sdFacesDataset.data.shape)
eigenFaces(sdFacesDataset)
```

```python
#Scaling Down Images
x,y = 32,32
suFacesDataset = facesDataset
suFacesDataset.images = resizeImages(facesDataset,x,y)
suFacesDataset.data = suFacesDataset.images.flatten().reshape(facesDataset.data.shape[0],x*y)
print(suFacesDataset.data.shape)
eigenFaces(suFacesDataset)
```
