```python
from PIL import Image
import matplotlib.pyplot as plot
import numpy as np
```

```python
def displayImage(image, title="No title"):
    plot.figure()
    plot.imshow(image)
    plot.title(title)
```

```python
def imgSVD(sample_img):
    #Extracting Pixels matrix from original image ------------------------------------------
    curr_band = 1
    pixelMatrix = np.matrix(list(sample_img.getdata(band=curr_band)), float)
    pixelMatrix.shape = (sample_img.size[1], sample_img.size[0])
    #displayImage(pixelMatrix, 'Band=' + str(curr_band) + ' Pixel Matrix')
    # --------------------------------------------------------------------------------------
    # print("Original image:\n", pixelMatrix)

    #Converting to grayscale image
    # gray_img = sample_img.convert('LA')
    # plot.figure(figsize=(9,6))
    # plot.imshow(gray_img)

    #Extracting Pixels matrix from Grayscaled image
    # pixelMatrix = np.matrix(list(gray_img.getdata(band=0)), float)
    # pixelMatrix.shape = (gray_img.size[1], gray_img.size[0])
    # print("Grayscale image:\n", pixelMatrix)
    #Plotting the grayscale image

    #SVD decomposition of original image -----------------------------------------------------
    U,S,V = np.linalg.svd(pixelMatrix)
    totalComps = U[0].size
    #U,S,V are all 1200x1200 in size
    return [U,S,V]
```

```python
def combPartialEigenComp(Ulist,Slist,Vlist,totalComps, interval):
    #Testing to see how the image looks like with partial number of eigen components -------------------
    figure = plot.figure(figsize = (50,50))
    i = 1
    totalCase = len(Ulist)
    for comps in range(1, totalComps+1, interval):
        for j in range(totalCase):
            PixelMatrix = np.matrix(Ulist[j][:,:comps]*np.diag(Slist[j][:comps])*np.matrix(Vlist[j][:comps,:]))
            img = figure.add_subplot(totalComps,totalCase,i, xticks=[], yticks=[])
            i+=1
            img.imshow(PixelMatrix)
```

```python
def eachEigenComp(Ulist,Slist,Vlist,totalComps):
    #Checking information associated with each one of the eigen componenets
    figure = plot.figure(figsize = (50,50))
    i = 1
    totalCase = len(Ulist)
    for comp_number in range(0,totalComps):
        for j in range(len(Ulist)):
            PixelMatrix = np.matrix(Ulist[j][:,comp_number]*Slist[j][comp_number]*Vlist[j][comp_number,:])
            img = figure.add_subplot(totalComps,totalCase,i, xticks=[], yticks=[])
            i+=1
            img.imshow(PixelMatrix)
```

```python
sample_img = Image.open('sample_img_2.jpg')
# displayImage(sample_img, 'Sample Image')
U,S,V = imgSVD(sample_img)

#Scaling Up the image
scaled_up_img = sample_img.resize((2000,2000))
U_up, S_up, V_up = imgSVD(scaled_up_img)

#Scaliing Down the image
scaled_down_img = sample_img.resize((600,600))
U_down, S_down, V_down = imgSVD(scaled_down_img)


Ulist = [U,U_up,U_down]
Vlist = [V,V_up,V_down]
Slist = [S,S_up,S_down]
# print("Sample Image object:\n", sample_img)

# --------------------------------------------------------------------------------------
# print(U.shape,pixelMatrix.shape)
# print(pixelMatrix)
combPartialEigenComp(Ulist,Slist,Vlist,5, 1)
#By 20 components, we are getting a clear image
# ---------------------------------------------------------------------------------------------------
```

```python
eachEigenComp(Ulist,Slist,Vlist,10)
```

```python
#Rotating the Image
rotated_img = sample_img.rotate(90)
U_r, S_r, V_r = imgSVD(rotated_img)
Ulistr = [U, U_r]
Vlistr = [V, V_r]
Slistr = [S, S_r]
combPartialEigenComp(Ulistr,Slistr,Vlistr,5, 1)
```

```python
eachEigenComp(Ulistr,Slistr,Vlistr,10)
```
